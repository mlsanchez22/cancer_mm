```r
create_file_launcher_custom_circuits_male <- function(project_name) {
  filename <- paste0(
    "/mnt/lustre/scratch/CBRA/research/projects/tcga_mm/code/09_drug_repository/09_6_drexml_tcga/",
    project_name,
    "/launcher_custom_circuits_male.sh"
  )
  
  script <- paste(
    "#!/bin/bash\n",
    "#SBATCH -J dr-schizo-launcher\n",
    "#SBATCH -t 0\n",
    "#SBATCH --mem 8G\n",
    "#SBATCH -n 1\n",
    "#SBATCH -o pipeline_custom_circuits.log\n\n",
    "###################################################\n",
    "# SET THE FOLLOWING VARIABLES TO MATCH YOUR NEEDS #\n",
    "###################################################\n",
    "SINGULARITY_MOD='Singularity/3.9.7'\n",
    "MAMBA_MOD='Mamba/4.14.0-0'\n",
    "PARAMS_FILE='/mnt/lustre/scratch/CBRA/research/projects/tcga_mm/code/09_drug_repository/09_6_drexml_tcga/",
    project_name,
    "/params_custom_circuits_male.yml'\n",
    "CONDA_INIT_SCRIPT='/mnt/lustre/expanse/software/easybuild/Rocky/8.5/Skylake/software/Mamba/4.14.0-0/etc/profile.d/conda.sh'\n",
    "CONDA_ENV='nextflow'\n",
    "UPDATE_NEXTFLOW='TRUE'\n\n",
    "# The PIPELINE_REPO variable must point to the github repository (user/project) or the main.nf full path (in your local -cloned- repository)\n",
    "#PIPELINE_REPO='babelomics/dr-pipeline'\n",
    "PIPELINE_REPO='/mnt/lustre/scratch/CBRA/research/projects/tcga_mm/code/09_drug_repository/dr-pipeline/main.nf'\n\n",
    "########\n",
    "# INIT #\n",
    "########\n",
    "# check params file\n",
    "[[ -f \"${PARAMS_FILE}\" ]] || { echo \"Params file ${PARAMS_FILE} not found!\" && exit 1; }\n\n",
    "# check if singularity module is loaded\n",
    "[[ $(module is-loaded ${SINGULARITY_MOD}) ]] || module load ${SINGULARITY_MOD}\n\n",
    "# check if mamba module is loaded\n",
    "[[ $(module is-loaded ${MAMBA_MOD}) ]] || module load ${MAMBA_MOD}\n\n",
    "# initialize conda\n",
    "source ${CONDA_INIT_SCRIPT} || { echo \"The conda environment couldn't be initialized!\" && exit 1; }\n\n",
    "# check conda env\n",
    "if [[ ! $(conda info --envs |grep ${CONDA_ENV}) ]]\n",
    "then\n",
    "  echo \"${CONDA_ENV} environment not found. Trying to create a new conda environment...\"\n",
    "  conda create -y -n ${CONDA_ENV} bioconda::nextflow\n",
    "fi\n\n",
    "# activate the conda environment\n",
    "conda activate ${CONDA_ENV} || { echo \"The conda environment couldn't be activated!\" && exit 1; }\n\n",
    "# update nextflow\n",
    "[[ \"${UPDATE_NEXTFLOW}\" ]] && nextflow self-update\n\n",
    "########\n",
    "# MAIN #\n",
    "########\n",
    "# run the workflow from the github repository\n",
    "nextflow run ${PIPELINE_REPO} -profile slurm,singularity -resume -params-file ${PARAMS_FILE}\n",
    sep=""
  )
  
  write(script, file = filename)
}


lapply(proj_tumorvsctrl$tcga, FUN=function(x) create_file_launcher_custom_circuits_male(x))
```
