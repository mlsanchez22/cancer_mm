```r
create_circ_file_funct <- function(project){
    circs_sign_male <- unique(circs_sig_dif[circs_sig_dif$project == project & circs_sig_dif$sex == "male","circ_hip"]) 
    circs_no_sign_male <- setdiff(circs, circs_sign_male)
    circs_drugrep_male <- data.frame("index"=c(circs_no_sign_male, circs_sign_male),"in_disease"=c(rep(0,length(circs_no_sign_male)), rep(1,length(circs_sign_male))))
    write.table(circs_drugrep_male, file = paste("/mnt/lustre/scratch/CBRA/research/projects/tcga_mm/results/09_drug_repository/09_2_circuits_sign_limma_tumvsctrl/circuits_male_",project,".tsv",sep=""), row.names=FALSE, sep="\t",quote=F)

    circs_sign_female <- unique(circs_sig_dif[circs_sig_dif$project == project & circs_sig_dif$sex == "female","circ_hip"]) 
    circs_no_sign_female <- setdiff(circs, circs_sign_female)
    circs_drugrep_female <- data.frame("index"=c(circs_no_sign_female, circs_sign_female),"in_disease"=c(rep(0,length(circs_no_sign_female)), rep(1,length(circs_sign_female))))
    write.table(circs_drugrep_female, file = paste("/mnt/lustre/scratch/CBRA/research/projects/tcga_mm/results/09_drug_repository/09_2_circuits_sign_limma_tumvsctrl/circuits_female_",project,".tsv",sep=""), row.names=FALSE, sep="\t",quote=F)
}
```
